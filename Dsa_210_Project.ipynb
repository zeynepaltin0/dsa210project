```python
#Install the Kaggle package
!pip install -q kaggle

from google.colab import files
files.upload()

!mkdir -p ~/.kaggle
!mv 'kaggle(2).json' ~/.kaggle/kaggle.json
!chmod 600 ~/.kaggle/kaggle.json

!kaggle datasets download -d unsdsn/world-happiness -p /content

!unzip -o /content/world-happiness.zip -d /content

import pandas as pd
df_2016 = pd.read_csv('/content/2016.csv')

df_2016.head()
```

```python
#Check for missing data in the 2016 dataset
missing_data_2016 = df_2016.isnull().sum()

print("Missing data for 2016 dataset:")
print(missing_data_2016)
```

```python
# Drop the confidence interval columns from df_2016
df_2016.drop(['Lower Confidence Interval', 'Upper Confidence Interval'], axis=1, inplace=True)

print("Updated columns in df_2016:")
print(df_2016.columns)
```

```python
# Merge the data sets
df_suicide_filtered.rename(columns={'country': 'Country'}, inplace=True)

df_merged_2015 = pd.merge(df_suicide_filtered[df_suicide_filtered['year'] == 2015], df_2015, on=['Country'], how='left')
df_merged_2016 = pd.merge(df_suicide_filtered[df_suicide_filtered['year'] == 2016], df_2016, on=['Country'], how='left')

df_merged = pd.concat([df_merged_2015, df_merged_2016])

df_merged_cleaned = df_merged.dropna(subset=['Happiness Score'])

print(f"Cleaned merged dataset:\n{df_merged_cleaned.head()}")

print(f"\nCleaned merged dataset shape: {df_merged_cleaned.shape}")
```

```python
# Check the columns of the datasets
print(f"Columns in df_suicide_filtered:\n{df_suicide_filtered.columns}")
print(f"\nColumns in df_2015:\n{df_2015.columns}")
print(f"\nColumns in df_2016:\n{df_2016.columns}")

print(f"\nColumns in df_merged_cleaned:\n{df_merged_cleaned.columns}")

```

```python
# Compute correlation with Happiness Score
happiness_components = [
    'Economy (GDP per Capita)',
    'Health (Life Expectancy)',
    'Freedom',
    'Trust (Government Corruption)',
    'Generosity',
    'Dystopia Residual'
]

df_merged_cleaned[happiness_components + ['Happiness Score']].corr()['Happiness Score'].sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import statsmodels.api as sm
```

```python
#Distribution of suicide rate (per 100k) (Histogram)
plt.figure(figsize=(8, 5))
sns.histplot(df_merged_cleaned['suicides/100k pop'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Suicide Rates (per 100k)')
plt.xlabel('Suicide Rate per 100k')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Distribution of hapiness scores (Histogram)
plt.figure(figsize=(8, 5))
sns.histplot(df_merged_cleaned['Happiness Score'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Hapiness Score vs Suicide Rate (with country)
country_avg = df_merged_cleaned.groupby('Country').agg({
    'Happiness Score': 'mean',
    'suicides/100k pop': 'mean'
}).reset_index()

country_avg_sorted = country_avg.sort_values(by='Happiness Score', ascending=False)

plt.figure(figsize=(14, 8))
sns.barplot(
    data=country_avg_sorted,
    x='Country',
    y='suicides/100k pop',
    palette='coolwarm'
)
plt.xticks(rotation=90)
plt.title('Average Suicide Rate by Country Ranked by Happiness Score')
plt.xlabel('Country (Ordered by Happiness Score)')
plt.ylabel('Average Suicides per 100k Population')
plt.tight_layout()
plt.show()
```

```python
#Hapiness Score vs Suicide Rate Plot
plt.figure(figsize=(10, 6))
sns.regplot(
    data=df_merged_cleaned,
    x='Happiness Score',
    y='suicides/100k pop',
    scatter_kws={'alpha': 0.5, 'color': 'blue'},
    line_kws={'color': 'red', 'linestyle': '--'}
)
plt.title('Happiness Score vs. Suicide Rate (per 100k pop)')
plt.xlabel('Happiness Score (Higher = Happier)')
plt.ylabel('Suicides per 100k Population')
plt.grid(True)
plt.show()
```

```python
#With reigons
sns.lmplot(data=df_merged_cleaned, x='Happiness Score', y='suicides/100k pop', hue='Region', height=6, aspect=1.5, scatter_kws={'alpha':0.5})
plt.title('Regression: Happiness Score vs. Suicide Rate')
plt.xlabel('Happiness Score')
plt.ylabel('Suicides per 100k population')
plt.grid(True)
plt.show()
```

```python
#Differentiated by se
g = sns.lmplot(data=df_merged_cleaned, x='Happiness Score', y='suicides/100k pop',
               col='sex', hue='sex', height=6, aspect=1)
g.set_titles('Sex: {col_name}')
plt.suptitle('Happiness vs Suicide Rate by Sex', y=1.03)
plt.show()
```

```python
#Boxplot by sex
plt.figure(figsize=(6, 5))
sns.boxplot(x='sex', y='suicides/100k pop', data=df_merged_cleaned, palette='Set2')
plt.title('Suicide Rate per 100k by Sex')
plt.xlabel('Sex')
plt.ylabel('Suicides per 100k Population')
plt.grid(True)
plt.show()
```

```python
df_male = df_merged_cleaned[df_merged_cleaned['sex'] == 'male']
df_female = df_merged_cleaned[df_merged_cleaned['sex'] == 'female']

# Calculate Pearson correlation for males
r_male, p_male = pearsonr(df_male['Happiness Score'], df_male['suicides/100k pop'])

# Calculate Pearson correlation for females
r_female, p_female = pearsonr(df_female['Happiness Score'], df_female['suicides/100k pop'])

print(f"Male: Pearson r = {r_male:.4f}, p-value = {p_male:.4f}")
print(f"Female: Pearson r = {r_female:.4f}, p-value = {p_female:.4f}")
```

```python
#Heatmap
sns.heatmap(df_merged_cleaned[['suicides/100k pop', 'Happiness Score', 'Family', 'Health (Life Expectancy)', 'gdp_per_capita ($)']].corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Regression Analysis
X = df_merged_cleaned['Happiness Score']
y = df_merged_cleaned['suicides/100k pop']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
#Multiple Regression Analysis
X = df_merged_cleaned[['Happiness Score', 'gdp_per_capita ($)', 'Family', 'Health (Life Expectancy)']]
X = sm.add_constant(X)
y = df_merged_cleaned['suicides/100k pop']

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
#Hypothesis Testing
correlation, p_value = pearsonr(df_merged_cleaned['Happiness Score'], df_merged_cleaned['suicides/100k pop'])
print("Correlation:", correlation)
print("P-Value:", p_value)
if p_value < 0.05:
    print("Significant relationship.")
else:
    print("No significant relationship.")
```
